```python
#import libs
from ultralytics import YOLO
import os
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# Load the YOLO model for pose estimation
model = YOLO('yolov8n-pose.pt')
```

```python
#data we want to use to predict the pose
img_test_folder='/content/drive/MyDrive/imgs_test_set'
imgs=[i for i in os.listdir(img_test_folder) if i.endswith(('.jpg', '.jpeg', '.png'))]
```

```python
for img in imgs:
  #Full path
  img_full_path=os.path.join(img_test_folder,img)
  # Predict with the model
  results = model(img_full_path, save=True)
  print(f"Processed {img}")
```

```python
output_folder='/content/drive/MyDrive/imgs_test_set/runs/pose/predict'
```

```python
# Get a list of all image files in the folder
image_files = [f for f in os.listdir(output_folder) if f.endswith(('.jpg', '.jpeg', '.png'))]
for image_file in image_files:
    # Construct the full path to the image
    image_path = os.path.join(output_folder, image_file)

    # Load and display the image
    img = mpimg.imread(image_path)
    plt.imshow(img)
    plt.axis('off')  # Hide axis
    plt.title(image_file)  # Display the image filename as the title
    plt.show()
```
